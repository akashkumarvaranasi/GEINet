```lua
datapath = "/home/chenqiang/data/gait-rnn"
modelname = "attention"
geipath = "/home/chenqiang/data/OULP_C1V1_Pack_GEI"
local prepDataset = require 'prepareDataset'
dataset = prepDataset.prepareDatasetOULP(datapath, modelname, geipath)
```

```lua
bsize = 4
geis = {}
for _, t in ipairs({"train", "val", "test"}) do
    t_batch = dataset[t]:next_batch(bsize)
    t_geis = torch.randn(4, 1, 128, 88)
    for i=1,bsize do
        t_geis[{{i,}}] = t_batch[i][1]
    end
    geis[t] = t_geis
end
```

```lua

for _, t in ipairs({"train", "val", "test"}) do
    itorch.image(geis[t])
end
```
